### Visualization

```python
toxic_comments = train[train.toxic == 1].comment_text.values
```

```python
figure = plt.figure(figsize=(10,10))

wordcloud = WordCloud(stopwords = STOPWORDS).generate(str(toxic_comments))

plt.imshow(wordcloud, interpolation='bilinear')
plt.tight_layout()
plt.axis('off')
plt.show()
```

```python
train.toxic.value_counts().plot(kind='bar')
plt.title('Distribution of toxic and non-toxic comments')
plt.xticks(ticks=[0, 1], labels=['non-toxic','toxic'], rotation=0)
plt.ylabel('count')
plt.show()
```
